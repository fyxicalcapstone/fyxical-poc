```python
from google.cloud import bigquery
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

PROJECT_ID = "fyxical-poc"
DATASET    = "fyxical_poc"
client     = bigquery.Client(project=PROJECT_ID)

patients = client.query(f"SELECT * FROM `{PROJECT_ID}.{DATASET}.patient_intake`").to_dataframe()
print(f"Patients loaded: {len(patients)}")
print(patients.describe().round(2).to_string())
```

```python
# Acronym Reference

ACRONYMS_CLINICAL = {
    "OSW" : "Oswestry Disability Index — 0-100% disability scale. "
             "Minimal 0-20% / Moderate 21-40% / Severe 41-60% / "
             "Crippled 61-80% / Bed-bound 81-100%.",
    "VAS" : "Visual Analogue Scale — pain score 0-10. "
             "Mild 0-3 / Moderate 4-6 / Severe 7-10.",
    "EDA" : "Exploratory Data Analysis — descriptive statistics and "
             "visualisations to understand data distributions before modelling.",
}

ACRONYMS_PATHWAYS = {
    "ACUTE_MILD_MOD" : "Acute + Mild/Moderate  — Pain score 0-6, Duration <2 weeks",
    "ACUTE_SEVERE"   : "Acute + Severe         — Pain score 7-10, Duration <2 weeks",
    "SUBACUTE"       : "Subacute               — Duration 2-12 weeks, any severity",
    "CHRONIC"        : "Chronic / High Burden  — Duration >12 weeks OR high OSW disability",
}

print("=" * 65)
print("  ACRONYM REFERENCE — Notebook 02")
print("=" * 65)

print("\n── Clinical & Project Terms ────────────────────────────────")
for code, desc in ACRONYMS_CLINICAL.items():
    print(f"\n  {code}")
    words = desc.split()
    line = "    "
    for word in words:
        if len(line) + len(word) + 1 > 64:
            print(line)
            line = "    " + word + " "
        else:
            line += word + " "
    print(line.rstrip())

print("\n── Pathway Codes ───────────────────────────────────────────")
for code, desc in ACRONYMS_PATHWAYS.items():
    print(f"  {code:<16} {desc}")

print("\n" + "=" * 65)
```

```python
# Clinical variable distributions

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

patients["pain_severity_score_0_10"].hist(ax=axes[0], bins=10, color="#4a7c9e", edgecolor="white")
axes[0].set_title("Pain Severity Score (0-10)")
axes[0].set_xlabel("Score"); axes[0].set_ylabel("Count")

patients["osw_baseline_percent_disability"].hist(ax=axes[1], bins=10, color="#6a9e6a", edgecolor="white")
axes[1].set_title("OSW Baseline Disability (%)")
axes[1].set_xlabel("Disability %")

patients["pain_duration_weeks"].hist(ax=axes[2], bins=15, color="#9e7a4a", edgecolor="white")
axes[2].set_title("Pain Duration (weeks)")
axes[2].set_xlabel("Weeks")

plt.suptitle("Clinical Variable Distributions", fontweight="bold", y=1.02)
plt.tight_layout()
plt.savefig("clinical_distributions.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# Clinical category distributions

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

patients["pain_duration_category"].value_counts().plot(
    kind="bar", ax=axes[0], color="#4a7c9e", edgecolor="white")
axes[0].set_title("Pain Duration Category")
axes[0].set_xlabel("Acute <2wks / Subacute 2-12wks / Chronic >12wks")
axes[0].tick_params(axis="x", rotation=30)

patients["pain_severity_category"].value_counts().plot(
    kind="bar", ax=axes[1], color="#6a9e6a", edgecolor="white")
axes[1].set_title("Pain Severity Category")
axes[1].set_xlabel("Mild 0-3 / Moderate 4-6 / Severe 7-10")
axes[1].tick_params(axis="x", rotation=30)

osw_order = ["Minimal","Moderate","Severe","Crippled","Bed-bound"]
patients["osw_baseline_disability_category"].value_counts().reindex(osw_order, fill_value=0).plot(
    kind="bar", ax=axes[2], color="#9e6a9e", edgecolor="white")
axes[2].set_title("OSW Disability Category")
axes[2].set_xlabel("Minimal 0-20% ... Bed-bound 81-100%")
axes[2].tick_params(axis="x", rotation=30)

plt.suptitle("Clinical Category Distributions", fontweight="bold", y=1.02)
plt.tight_layout()
plt.savefig("clinical_categories.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# Demographic breakdown

fig, axes = plt.subplots(2, 2, figsize=(14, 9))

patients["sex"].value_counts().plot(kind="bar", ax=axes[0,0], color=["#4a7c9e","#9e6a9e"], edgecolor="white")
axes[0,0].set_title("Sex Distribution"); axes[0,0].tick_params(axis="x", rotation=0)

patients["race_ethnicity"].value_counts().plot(kind="barh", ax=axes[0,1], color="#6a9e6a", edgecolor="white")
axes[0,1].set_title("Race/Ethnicity Distribution")

patients["occupation_type"].value_counts().plot(kind="barh", ax=axes[1,0], color="#9e7a4a", edgecolor="white")
axes[1,0].set_title("Occupation Type")

patients["region_census"].value_counts().plot(kind="bar", ax=axes[1,1], color="#4a8a7e", edgecolor="white")
axes[1,1].set_title("Census Region"); axes[1,1].tick_params(axis="x", rotation=30)

plt.suptitle("Demographic Distributions (Equity Monitoring Variables)", fontweight="bold", y=1.02)
plt.tight_layout()
plt.savefig("demographics.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# Demographic Rationale

print("""
Demographic Variables: Purpose and Rationale
----------------------------------------------
Included for equity monitoring.

How they ARE used:
  - Descriptive analysis: understand who is represented in the dataset
  - Equity metrics: check whether pathway assignments differ
    systematically across groups with equivalent clinical profiles
  - Subgroup review: flag disproportionate pathway assignments

How they are NOT used:
  - Excluded from the ML feature set
  - Do NOT influence rule-based pathway assignment
  - Post-hoc analysis variables only

Clinical rationale:
  Two patients with identical OSW %, pain score, and duration
  should receive the same pathway regardless of demographics.
  Any systematic difference warrants clinical review.
""")
```

```python
# Clinical correlations

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(patients["pain_severity_score_0_10"],
                patients["osw_baseline_percent_disability"],
                alpha=0.4, color="#4a7c9e", edgecolors="none", s=30)
axes[0].set_title("Pain Score vs OSW Disability %", fontweight="bold")
axes[0].set_xlabel("Pain Severity Score (0-10)")
axes[0].set_ylabel("OSW Baseline Disability (%)")
corr1 = patients[["pain_severity_score_0_10","osw_baseline_percent_disability"]].corr().iloc[0,1]
axes[0].annotate(f"r = {corr1:.2f}", xy=(0.05, 0.92), xycoords="axes fraction", fontsize=11)

axes[1].scatter(patients["pain_duration_weeks"],
                patients["osw_baseline_percent_disability"],
                alpha=0.4, color="#6a9e6a", edgecolors="none", s=30)
axes[1].set_title("Pain Duration vs OSW Disability %", fontweight="bold")
axes[1].set_xlabel("Pain Duration (weeks)")
axes[1].set_ylabel("OSW Baseline Disability (%)")
corr2 = patients[["pain_duration_weeks","osw_baseline_percent_disability"]].corr().iloc[0,1]
axes[1].annotate(f"r = {corr2:.2f}", xy=(0.05, 0.92), xycoords="axes fraction", fontsize=11)

plt.tight_layout()
plt.savefig("clinical_correlations.png", dpi=150, bbox_inches="tight")
plt.show()
```
